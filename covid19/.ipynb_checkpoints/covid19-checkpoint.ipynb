```python
import pandas as pd
import seaborn as sb
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
data_frame = pd.read_csv('data.csv')
data_frame
```

```python
data_frame.describe()
```

```python
df1=pd.DataFrame(data_frame, columns=['ADM0_NAME','cum_conf','cum_death'])
sb.set()
graph = sb.scatterplot(x = 'cum_conf', y = 'cum_death', data = df1)
graph
```

```python
TOOLTIPS =HoverTool(tooltips=[
    ("index","$index"),
    ("(cum_conf,cum_death)","$cum_conf,$cum_death"),
    ("ADM0_NAME","@ADM0_NAME")]
)
p = figure(title="COVID19", x_axis_label = 'cum_conf', 
           y_axis_label = 'cum_death', plot_width=700, 
           plot_height=700, tools=[TOOLTIPS])
show(p)
```
